```python
import csv
import pandas as pd
```

```python
# Load the CSV file into a pandas DataFrame
file_path = "../Data/updated_movies2.csv"
movies_df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
movies_df.head()
```

```python
# Convert columns with numerical values to numeric type (if necessary) and calculate statistics
numerical_columns = ["opening_weekend", "percent_of_total_gross", "production_budget"]
movies_df[numerical_columns] = (
    movies_df[numerical_columns]
    .replace("[\$,%,]", "", regex=True)
    .apply(pd.to_numeric, errors="coerce")
)

# Calculate statistics
stats = (
    movies_df[numerical_columns]
    .agg(["min", "max", "mean", "median", "std"])
    .transpose()
)
stats["missing_values"] = movies_df[numerical_columns].isnull().sum()

# Display the statistics
print(stats)
```
